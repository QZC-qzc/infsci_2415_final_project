```python
import pandas as pd # for creating data frame
import numpy as np # for mathematical linear algebra
import matplotlib.pyplot as plt # for plotting data
import seaborn as sns
df = pd.read_csv('shopping_trends.csv')
sns.boxplot(data=df,x='Frequency of Purchases',y='Purchase Amount (USD)')
```

```python
total_purchase_by_season_frequency = df.groupby(['Frequency of Purchases', 'Season'])['Purchase Amount (USD)'].sum().unstack()

total_purchase_by_season_frequency = total_purchase_by_season_frequency.reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='Frequency of Purchases', y='value', hue='Season', data=total_purchase_by_season_frequency.melt(id_vars='Frequency of Purchases'))

plt.title('Total Purchase Amount for Different Seasons Across Each Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Season')
plt.show()
```

```python
def age_category(age):
    if age <= 18:
        return '0-18'
    elif age <= 60 and age > 18:
        return '18-60'
    else:
        return 'Above 60'

df['Age Group'] = df['Age'].apply(age_category)

total_purchase_by_agegroup_frequency = df.groupby(['Frequency of Purchases', 'Age Group'])['Purchase Amount (USD)'].sum().unstack()

total_purchase_by_agegroup_frequency = total_purchase_by_agegroup_frequency.reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='Frequency of Purchases', y='value', hue='Age Group', data=total_purchase_by_agegroup_frequency.melt(id_vars='Frequency of Purchases'))

plt.title('Total Purchase Amount for Different Age Groups Across Each Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.legend(title='Age Group')
plt.show()
```
